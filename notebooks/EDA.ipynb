```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import (accuracy_score,classification_report,confusion_matrix)
```

```python
def evaluate_model(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)

    print("="*60)
    print(model.__class__.__name__)
    print("="*60)

    print("\nAccuracy:")
    print(accuracy_score(y_test,y_pred))

    print("\nClassificationReport")
    print(classification_report(y_test,y_pred))

    cm=confusion_matrix(y_test,y_pred)
    
    plt.figure(figsize=(5,4))
    sns.heatmap(cm,annot=True,fmt="d",cmap="Blues")
    plt.title(model.__class__.__name__)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()


    
```

```python
from sklearn.tree import DecisionTreeClassifier
evaluate_model(DecisionTreeClassifier(random_state=42),X_train,X_test,y_train,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

evaluate_model(RandomForestClassifier(n_estimators=100,random_state=42),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
xgb=XGBClassifier(random_state=42,eval_metric="logloss")
```

```python
param_grid={"n_estimators":[50,100,150], "max_depth":[3,5,7],"learning_rate":[0.01,0.1,0.3]}
```
